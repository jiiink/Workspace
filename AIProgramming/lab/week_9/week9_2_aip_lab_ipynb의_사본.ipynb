# Helper Functions

* visualize_graph provides a function to visually represent a graph using the Graph class from the graphviz package.

* The function takes an adjacency list representation of a graph as its input, where each key-value pair corresponds to a node and its list of neighbors.

* The function then processes this adjacency list to construct a visual representation using nodes and edges.

````python
from graphviz import Graph

def visualize_graph(graph):
    dot = Graph(strict=True)

    for node in graph:
        dot.node(node)

    for node, neighbors in graph.items():
        for neighbor in neighbors:
            dot.edge(node, neighbor)

    return dot

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'C']
}

dot = visualize_graph(graph)
dot
````
